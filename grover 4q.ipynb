```python
from logicqubit.logic import *
import math
```

```python
def get_k(N):
    return round(pi*math.sqrt(N)/4)

get_k(16)
```

```python
logicQuBit  = LogicQuBit(5)

q1 = Qubit()
q2 = Qubit()
q3 = Qubit()
q4 = Qubit()
a = Qubit() # ancilla qubit

q1.H()
q2.H()
q3.H()
q4.H()
```

```python
# |0010> -> X1*X2*I*X4|q1>|q2>|q3>|q4>
q1.X()
q2.X()
#q3.X()
q4.X()

# CCCZ
a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q1.X()
q2.X()
#q3.X()
q4.X()

# |0110> -> X1*I*I*X4|q1>|q2>|q3>|q4>
q1.X()
#q2.X()
#q3.X()
q4.X()

# CCCZ
a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q1.X()
#q2.X()
#q3.X()
q4.X()

# Grover 1x
q1.H()
q2.H()
q3.H()
q4.H()
q1.X()
q2.X()
q3.X()
q4.X()

a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q4.X()
q3.X()
q2.X()
q1.X()
q4.H()
q3.H()
q2.H()
q1.H()
```

```python
logicQuBit.PlotDensityMatrix()
```

```python
# |0010> -> X1*X2*I*X4|q1>|q2>|q3>|q4>
q1.X()
q2.X()
#q3.X()
q4.X()

# CCCZ
a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q1.X()
q2.X()
#q3.X()
q4.X()

# |0110> -> X1*I*I*X4|q1>|q2>|q3>|q4>
q1.X()
#q2.X()
#q3.X()
q4.X()

# CCCZ
a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q1.X()
#q2.X()
#q3.X()
q4.X()

# Grover 2x
q1.H()
q2.H()
q3.H()
q4.H()
q1.X()
q2.X()
q3.X()
q4.X()

a.CCX(q1, q2)
q4.H()
q4.CCX(a, q3)
q4.H()
a.CCX(q1, q2)

q4.X()
q3.X()
q2.X()
q1.X()
q4.H()
q3.H()
q2.H()
q1.H()
```
