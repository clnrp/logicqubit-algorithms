```python
from logicqubit.logic import *
import random
import numpy as np
```

```python
logicQuBit  = LogicQuBit(5)

A = Qubit()
B = Qubit()
Psi = Qubit()
x1 = Qubit()
x2 = Qubit()

# prepare psi state
x1.H()
x2.H()
Psi.CCX(x1,x2) # and port

# generate entangled state
A.H()
B.CNOT(A)

# teleportation operations
A.CNOT(Psi)
Psi.H()
```

```python
# classical information and adjustment operations
value = logicQuBit.Measure_One(A)
if value[0]==1:
    B.X()

value = logicQuBit.Measure_One(Psi)
if value[0]==1:
    B.Z()    
    
logicQuBit.PlotDensityMatrix()
```

```python
logicQuBit.Measure([B])
logicQuBit.Plot()
```
