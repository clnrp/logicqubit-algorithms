Estados de Bell
\begin{eqnarray*}
   |\Phi^{+}> = \frac{1}{\sqrt{2}}\left(|00>+|11> \right)
\end{eqnarray*}

\begin{eqnarray*}
   |\Phi^{-}> = \frac{1}{\sqrt{2}}\left(|00>-|11> \right)
\end{eqnarray*}

\begin{eqnarray*}
   |\Psi^{+}> =\frac{1}{\sqrt{2}}\left(|01>+|10> \right)
\end{eqnarray*}

\begin{eqnarray*}
   |\Psi^{-}> = \frac{1}{\sqrt{2}}\left(|01>-|10> \right)
\end{eqnarray*}

```python
logicQuBit  = LogicQuBit(5)

A = Qubit()
B = Qubit()
Psi = Qubit()
x1 = Qubit()
x2 = Qubit()

# prepare psi state
x1.H()
x2.H()
Psi.CCX(x1,x2) # and port

# generate entangled state
A.H()
B.CNOT(A)

# teleportation operations
A.CNOT(Psi)
Psi.H()
```

\begin{eqnarray*}
  |00>_{A} \rightarrow |\psi>_{B} = a|0>+b|1> = |\psi>_{A}\\
  |01>_{A} \rightarrow |\psi>_{B} = X(a|1>+b|0>) = |\psi>_{A} \\
  |10>_{A} \rightarrow |\psi>_{B} = Z(a\|0>-b|1>) = |\psi>_{A}\\
  |11>_{A} \rightarrow |\psi>_{B} = XZ(a|1>-b|0>) = |\psi>_{A} 
\end{eqnarray*}

```python
# classical information and adjustment operations
value = logicQuBit.Measure_One(A)
if value[0]==1:
    B.X()

value = logicQuBit.Measure_One(Psi)
if value[0]==1:
    B.Z()    
    
logicQuBit.PlotDensityMatrix()
```
